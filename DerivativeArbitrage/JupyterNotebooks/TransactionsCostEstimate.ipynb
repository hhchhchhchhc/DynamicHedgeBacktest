```python
coin = 'BTC'
import puffin.tardis_importer as ti
import pandas as pd
for day in range(14, 15):
    perp = ti.generate_tob_data_from_gzip(coin + '-PERP', 'C:/Users/Tibor/Data/tardis/ftx_quotes_2021-10-14_' + coin + '-PERP.csv.gz')
    spot = ti.generate_tob_data_from_gzip(coin + '/USD', 'C:/Users/Tibor/Data/tardis/ftx_quotes_2021-10-14_' + coin + '-USD.csv.gz')
    p = perp[['exchange_timestamp_nanos', 'bid_price', 'ask_price']].set_index('exchange_timestamp_nanos').rename(columns={'bid_price': 'perp_bid', 'ask_price': 'perp_ask'})
    s = spot[['exchange_timestamp_nanos', 'bid_price', 'ask_price']].set_index('exchange_timestamp_nanos').rename(columns={'bid_price': 'spot_bid', 'ask_price': 'spot_ask'})
    data = pd.merge(left=p, right=s, how='outer', left_index=True, right_index=True)
    data = data.sort_index()
    data = data.fillna(method='ffill')
    data = data.dropna(how='any')
    data['Perp_ask-Spot_bid'] = 2e4*(data['perp_ask'] - data['spot_bid'])/(data['perp_ask'] + data['spot_bid'])
    data['Perp_bid-Spot_ask'] = 2e4*(data['perp_bid'] - data['spot_ask'])/(data['perp_bid'] + data['spot_ask'])
    data['date'] = data.index // (1000000000*60*60*24)
    data['time'] = data.index % (1000000000*60*60*24)
    data['hour'] = data['time'] // (1000000000*60*60)
    stats = data[['date', 'hour', 'Perp_ask-Spot_bid', 'Perp_bid-Spot_ask']].groupby(['date','hour']).mean()
stats
```

```python
fundings = pd.read_parquet('C:/Users/Tibor/Sandbox/allfundings.parquet')
fundings['datetime'] = fundings.index
fundings['unix_seconds'] = fundings['datetime'].apply(lambda x: x.timestamp())
fundings['date'] = fundings['unix_seconds'].apply(lambda x: x // (60*60*24))
fundings['date'] = fundings['date'].astype(int)
fundings['unix_time'] = fundings['unix_seconds'].apply(lambda x: x % (60*60*24))
fundings['unix_time'] = fundings['unix_time'].astype(int)
fundings['hour'] = fundings['unix_time'].apply(lambda x: x // (60*60))
fundings['hour'] = fundings['hour'].astype(int)
fundings = fundings[['date', 'hour', coin + '-PERP/rate/funding']]
fundings = fundings.set_index(['date', 'hour'])
fundings[coin + '-PERP/rate/funding'] = fundings[coin + '-PERP/rate/funding']/365.25
fundings[coin + '-PERP/rate/funding'] = fundings[coin + '-PERP/rate/funding']*10000
fundings
```

```python
pd.merge(left=stats, right=fundings, how='left', left_index=True, right_index=True).plot()
```
